```python
import geopandas as gpd
```

```python
sidewalks = gpd.read_file("data/sidewalkwidths_nyc.geojson")
```

```python
sidewalks
```

```python
sidewalks.plot()
```

```python
zipcode = gpd.read_file("data/ZIP_CODE_040114.zip")
zipcode
```

```python
zipcode.plot()
```

```python
sidewalks.crs
```

```python
zipcode.crs
```

```python
zipcode = zipcode.to_crs(sidewalks.crs)
```

```python
zipcode.crs
```

```python
zipcode.plot()
```

```python
##spatial join

width = gpd.sjoin(sidewalks, zipcode, how = "left")
width
```

```python
width["id"].duplicated().value_counts()
```

```python
width[width["id"].duplicated()].plot()
```

```python
sidewalks[sidewalks["id"].duplicated()].plot()
```
